```
import pandas as pd
import folium
from folium.plugins import HeatMap
import branca
```

# Hora de processar

Na célula abaixo, vamos extrair os dados de geolocalização e separa-los em duas novas colunas. A coluna LAT armazenará a latitude, ao passo que a LNG conterá a longitude.

```
vetor = data.GeoLoc.str.split(',')

latitude = []
longitude = []
for loc in vetor:
  latitude.append( float( loc[0].strip() ) )
  longitude.append( float( loc[1].strip() ) )

data['LAT'] = pd.Series(latitude)
data['LNG'] = pd.Series(longitude)

data.head()
```

## Hora do show

Agora vamos criar um mapa coma biblioteca folium, para, em seguida, percorrer nosso dataset preenchendo os pontos no mapa.

```
m = folium.Map(
    location=[0, 0],
    zoom_start=2,
    tiles='Stamen Terrain'
)
```

```
for i in range(len(data)):
    folium.Marker([data.loc[i,'LAT'], data.loc[i,'LNG']],
                  icon=folium.Icon(
                          color = 'black',
                          icon = 'exclamation-triangle',
                          prefix='fa'),
                  popup = '<div style="text-align: center;">'+data.loc[i, 'Nome']+' - '+str(data.loc[i, 'Ano'])+'</div><hr><p>'+data.loc[i, 'Descricao']+'</p>'
        ).add_to(m)

m
```

No mapa acima, clique em qualquer ponto para ver uma breve descrição da batalha.

# Um datase mais completo

Através [desta fonte](https://www.kaggle.com/domcastro/wars-and-fights-draft/code) obtivemos acesso a um dataset muito mais completo com as batalhas ocorridas na segunda guerra.

Com mais de 1500 entradas este dataset é ideal para que possamos plotar um mapa de calor, indicando as áreas do globo afetadas por batalhas da segunda guerra.

```
dataCalor = pd.read_csv("Geodata.csv")
dataCalor.head()
```

```
coordinates_elem = []

for i in range(len(dataCalor)):
  if str(dataCalor.loc[i,'DecLatitude']) != "nan" and str(dataCalor.loc[i,'DecLongitude']) != "nan":
    coordinates_elem.append([dataCalor.loc[i,'DecLatitude'], dataCalor.loc[i,'DecLongitude']])

m_elem = folium.Map(
    location=[0, 0],
    zoom_start=2,
    tiles='Stamen Terrain'
)

HeatMap(coordinates_elem).add_to(m_elem)
```

```
f = branca.element.Figure()

f.add_child(m_elem)
```

## Aerial batles

Obtivemos também, um dataset com batalhas, especificamente aéreas, ocorridas na época, o mapa de calor abaixo mostra algumas áres onde ocorreram batalhas.

---

Infelizmente não é possível plotar todas as batalhas neste mapa, pois a bilbioteca não suporta o grande número de entradas deste dataset em específico.

```
operations = pd.read_csv("operations.csv")
operations.head()
```

O dataset "operations.csv" possui dados sobre as batalhas aereas que ocorreram durante a segunda guerra mundial. Logo abaixo o leitor pode ver que existem muitas colunas, os nomes são auto expressivos. Os mapas a frente vão tentar mostrar o local das batalhas e ao clicar na localização sera informado qual o pricipal esquadrão que estava na batalha e a data em que ocorreu. Por conta disso para garantir que não averá erros gerados pelo uso de valores "nan" então uma limpeza será feita antes de plotar o mapa.

```
# Abaixo estão todas as colunas disponiveis no data set

operations.columns
```

```
# A limpeza dos dados para retirar os valores "nan"

operations_clean = operations
operations_clean = operations_clean[ operations_clean['Target Longitude'].notnull()]
operations_clean = operations_clean[ operations_clean['Target Latitude'].notnull()]
operations_clean = operations_clean[ operations_clean['Mission Date'].notnull()]
operations_clean = operations_clean[ operations_clean['Air Force'].notnull()]
operations_clean = operations_clean[ operations_clean['Target Country'].notnull()]
```

```
# Criação objeto mapa folium que receberá os pontos

aerial_batle_points = folium.Map(
    location=[0, 0],
    zoom_start=2,
    tiles='Stamen Terrain'
)
```

```
#    Iteração de linha por linha do dataset para colocar
# no mapa do folium 1000 pontos de batalhas aéreas da
# segunda guerra mundial.

for index, row in operations_clean.iterrows():
  if index > 1000:
    break
  folium.Marker([row['Target Latitude'], row['Target Longitude']],
                icon=folium.Icon(color = 'black', icon = 'exclamation-triangle',prefix='fa'), 
                popup = '<div style="text-align: center;">'+str(row['Air Force'])+' - '+ str(row['Mission Date'])+'<br><p>alvo: '+ str(row['Target Country'])+'<br><p>Falhas em aviões:'+str(row['Aircraft Failed'])+'</p><br><p>Aviões avariados:'+ str(row['Aircraft Damaged'])+'</p><br><p>Aviões perdidos:'+ str(row['Aircraft Lost'])+'</p><br><p>Aviões retornados:'+ str(row['Aircraft Returned'])+'</p>'+ '</div>'
               ).add_to(aerial_batle_points)
```

  No mapa acima existem 1000 pontos de batalha aérea da segunda guerra mundial. Entretanto o dataset possui cerca de 119mil pontos após a limpeza de valores "nan". Ouve a tentativa de colocar todos os pontos no mapa, mas a biblioteca aparentemente quebra, provavelmente pois precisa de muita memória para plotar tantos pontos.
Logo foi decidido que apenas 1000 pontos deveriam ser plotados, pois mais que isso gera lentidão no processamento e acima de 2000 pontos nem mesmo o mapa aparecem.

  Também ouve uma tentativa de se adicionar na janela de informações de cada ponto o numero de aviões que falharam, perdidos, avariados ou os que retornaram sem problemas, entretanto como é possivel ver abaixo nesse dataset não existe uma intercessão desses dados, assim, por mais que sejam dados interresantes foi descidido deixar-los de fora.
  
  Mesmo com uma fração minuscula da quantidade total de dados podemos ver que a maior parte da patalhas foram na Europa. As batalhas da Ásia mostram um comportamento interresante. Os pontos formam uma concavidade ao redor do japão, com um "caminho" de pontos que leva da Nova Zelândia até a perte sul do Japão. Uma teoria que temos é que esse comportamento pode ser explicado pelas seguintes premissas:
  
  - O Japão foi um pais de eixo;
  - Os ataques contra o Japão são em grande maioria dos EUA;
  - Os EUA foi um dos paises que entraram ao lado dos **aliados**;
  - Uma extratégia militar para maximizar o dano contra o oponente e deixa-lo recuado é lutar contra o inimigo colocando-o no centro de uma concavidade, pois assim o inimigo terá muitos pontos para lutar contra e fica mais difícil de encontrar um angulo para avançar as tropas.
